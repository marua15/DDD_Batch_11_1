```python
from datasets import load_dataset, Dataset

dataset = load_dataset("IconicAI/DDD")
```

```python
from easynmt import EasyNMT

# Initialize the EasyNMT model
model = EasyNMT('opus-mt')

translated_dataset = []

# Iterate over the dataset and translate messages in the specified index range
for row in dataset["train"]:
    if 24200 <= row['index'] < 26400:
        print(f"Translating index {row['index']}")
        translated_row = row.copy()
        try:
            translated_message = model.translate(row['message'], target_lang='fr')
            translated_row['message'] = translated_message
        except Exception as e:
            print(f"Error translating message at index {row['index']}: {str(e)}")
        translated_dataset.append(translated_row)

# Create a new dataset with the translated rows
translated_dataset = Dataset.from_list(translated_dataset)
```

```python
# Create a new dataset with the translated rows
dataset = Dataset.from_list(translated_dataset)
```
